```python
from IPython.display import Image, display
from math import pi
import math
from io import BytesIO
import cairo

# Specify values for the colours
WHITE = (1,1,1)
BLACK = (0,0,0)
ONES = (183/256,110/256,35/256)
ZEROS = (201/256,204/256,183/256)
PADDING = (180/256,82/256,35/256)
```

```python
def draw_parachute(width=600, height=600, message=("DARE", "MIGHTY", "THINGS", ["N",118,10,31,"W",34,11,58])):
    if (isinstance(message, tuple) and len(message) == 4):
        surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, width, height)
        ctx = cairo.Context(surface)

        ctx.scale(width, height)
        ctx.set_line_width(0.001)
        
        finished_clock = 0
        max_clock = 80

        for index, r in enumerate([(0.5, message[3]), (0.45,None), (0.40, message[2]), (0.35, message[1]), (0.2, message[0])]):
            
            if index == 1:
                ctx.set_source_rgb(*BLACK)
                ctx.arc(0.5,0.5,r[0],0, 2*math.pi)
                ctx.fill_preserve()
                ctx.set_source_rgb(*BLACK)
                ctx.stroke()
            else:
        
                starting_clock = finished_clock
                ending_clock = (starting_clock + 80)%80
                current_clock=finished_clock

                def draw_slice(colour):
                    ctx.set_source_rgb(*colour)
                    ctx.move_to(0.5, 0.5)
                    ctx.arc(0.5,0.5,r[0],((2*math.pi)/80)*current_clock,((2*math.pi)/80)*(current_clock+1))
                    ctx.fill_preserve()
                    ctx.set_source_rgb(*BLACK)
                    ctx.stroke()

                def draw_gap():
                    nonlocal current_clock
                    for i in range(3):
                        draw_slice(SPACES)
                        current_clock += 1
                        current_clock = current_clock % 80

                def fill_circle():
                    nonlocal current_clock
                    nonlocal max_clock
                    while (current_clock+3)%80 != ending_clock:
                        draw_slice(PADDING)
                        current_clock+= 1
                        current_clock = current_clock % 80
                    draw_gap()



                for c in r[1]:
                    if isinstance(c, str):
                        binary = int.from_bytes(c.encode(), 'big')
                        binary -= 64
                    elif isinstance(c, int):
                        binary = c
                    binary_str = f"{binary:07b}"
                    for i in binary_str:
                        if i == '0':
                            draw_slice(ZEROS)
                        elif i == '1':
                            draw_slice(ONES)
                        else:
                            print(i)

                        current_clock += 1
                        current_clock = current_clock % 80

                    draw_gap()
                finished_clock = current_clock
                if current_clock != ending_clock:
                    fill_circle()

        with BytesIO() as fileobj:
            surface.write_to_png(fileobj)
            display(Image(fileobj.getvalue(), width=width+100, height=height+100))
```

```python
draw_parachute(message=("test", "test", "test", [9,9,9,9,9,9,9,9]))
draw_parachute()
```
